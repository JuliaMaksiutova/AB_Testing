```python
import numpy as np
import pandas as pd
import statsmodels.stats.proportion as smp
```

### Task 1

```python
alpha = 0.05
# Group 1
first_group_number = 15550
first_group_purchases = 164
#Group 2
second_group_number = 15550
second_group_purchases = 228
```

Обе переменные качественные.<br>
Выборка большая (> 30 значений).<br>
Данных мало (нет рядов).<br>
Поэтому для проверки предлагаю использовать Z-test.<br>

```python
successes = list([first_group_purchases, second_group_purchases])
nobs = np.array([first_group_number, second_group_number])
zstat, pvalue = smp.proportions_ztest(count=successes, nobs=nobs, value=0, alternative='two-sided')
print(zstat)
print(pvalue)
```
