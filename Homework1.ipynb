<b>Выбранный продукт - ozon.ru</b>
<div><br></div>Курсив
Ключевые метрики:<br>
«Ozon» - маркетплейс, в котором можно выделить следующие направления операционной деятельности:
<ul>
<li>Лэндинг (представление товаров магазина в рамках общего портала)</li>
<li>Логистика (хранение и доставка товаров от продавцов к покупателям)</li>
<li>Финансы (расчеты между продавцами и покупателями, предоставление финансирования OZON карта и т.д.)</li>
</ul><br>
В рамках настоящего задания будем оценивать операционное направление в наибольшей степени связанное с web-инфраструктурой - Лэндинг.<br>
Для целей оценки выделим две цепочки показателей, соответствующие двум группам результатов:
<ol><li>Покупатели.</li>
- покупатели в целом;<br>
- постоянные покупатели.
<li>Продвижение</li></ol>

<h5>Пользователи</h5>
Общее количество пользователей платформы. Наш главный актив. Потенциальный продавец регистрируясь на платформе строит все свои планы продаж отталкиваясь от этого числа как теоретического максимума. Обе цепочки (Покупатели и Продвижение будут иметь этот показатель в качестве начальной точки).
<li>Total user - общее количество пользователей.</li><br>

<h5>Покупатели в целом</h5>
Total user -> Returning User Rate -> Shopping Cart Abandonment Rate -> Conversion Rate<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;-> Bounce Rate
<br>
Что происходит с нашим ключевым активом. Какое его качество (Conversion Rate, Shopping Cart Abandonment Rate). На какие темпы роста мы можем рассчитывать в случае запуска каких-то рекламных компаний (Returning User Rate, Bounce Rate)?
<li>Returning User Rate - процент вернувшихся пользователей на сайт из общего количества посетителей. Пользователь мониторит в первую очередь цены и наличие (скорость доставки). В процессе мониторинга он смотрит несколько источников. Если он возвращается к нам, значит наши условия выгоднее. Не всеми такими параметрами мы можем управлять. Например, мы имеем лишь ограниченное влияние на цены продавцов, но можем наращивать ассортимент, привлекая больше продавцов и повышать скорость доставки, открывая больше партнерских точек выдачи.</li>
<li>Bounce Rate - процент пользователей, которые покинули сайт после просмотра одной страницы, не совершив никаких дальнейших действий. Если пользователь не захотел даже попытаться найти товар на нашем сайте, то это большая проблема. Это значит, что проблема в первую очередь в нас как портале. Покупателя могла испугать высокая цена продавца, тогда надо смотреть вернется ли он к нам (если цены на эту группу товара высокие в принципе) или не вернется (это значит, что у нас не достаточно продавцов этой группы товаров или не правильно настроена выдача)</li>
<li>Shopping Cart Abandonment Rate - процент пользователей, которые добавили товар в корзину, но не совершили покупку. Если пользователь добавил товар в корзину, значит морально он готов к покупке, ему помешали какие-то иные обстоятельства кроме желания обладать этой вещью. Наша задача понять, что именно помешало и решить эту проблему с минимальным риском для себя. Если это высокая цена - предложить ему через какое-то время скидку, если ему не хватает денег - кредит (наш или партнера) и т.д.</li>
<li>Conversion Rate - сколько пользователей совершили покупку. Конечная точка цепочки. Ключевой показатель. Потенциальный продавец при принятии решения о запуске продаж на платформе будет ориентироваться именно на него + наш доход включает процент с продаж.</li>

<h5>Постоянные покупатели</h5>
Total user -> Conversion Rate -> Customer Retention Rate<br>
Качество самой платформы. На сколько пользователям удобно сразу совершать покупки у нас даже без учета других ресурсов. Для повышения этих метрик необходимо расширять инфраструктуру и ассортимент.
<ul>
<li>Conversion Rate - сколько пользователей зарегистрировались</li>
<li>Customer Retention Rate - процент клиентов, которые возвращаются для совершения повторных покупок.</li>
</ul>

<h5>Продвижение</h5>
View -> Reach -> Click<br>
Как наша платформа выглядит на фоне других интернет-магазинов и маркет-плейсов.
Как часто выбирают пользователи наш маркетплейс из других.
Оценка текущего состояния и точек роста нашего базового актива (см. "Покупатели в целом").
<ul>
<li>View - количество показов рекламных объявлений.</li>
<li>Reach - сколько из общего количества показов объявления увидели уникальные пользователи.</li>
<li>Click - количество кликов по рекламным объявлениям.</li>
</ul>
