<u>Гипотеза 1:</u><br>

<p>Купить другой онлайн магазин (прецедент - приобретение X5 Болеро.ру URL: https://www.cnews.ru/news/top/vladelets_perekrestka_kupil_bolero_i_003_ru)</p>
<p><u>Что делаем в каждой из групп:</u><br>
Предлагаем пользователям группы "A" приобрести аналогичный товар на портале Ozon<br>
Предлагаем пользователям группы "B" приобрести аналогичный товар с единоразовой премией в баллах 5 % от общего баланса баллов и 15 % при условии офомрмления карты Ozon
</p>

Conversion Rate - сколько пользователей совершили покупку.<br>
</p>
</div>

</p>
